```python
import math
```

```python
# parameters
W_f, W_hf, b_f = 0.5, 0.1, 0.0
W_i, W_hi, b_i = 0.6, 0.2, 0.0
W_c, W_hc, b_c = 0.7, 0.3, 0.0
W_o, W_ho, b_o = 0.8, 0.4, 0.0
W_y, b_y = 4.0, 0.0

# states 
h = 0.0
C = 0.0
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def tanh_activation(x):
    return math.tanh(x)
```

```python
# Input
X = [1, 2, 3, 4] 

for t, x in enumerate(X, start=1):
    # forget gate
    f = sigmoid(W_f * x + W_hf * h + b_f)
    
    # input gate
    i = sigmoid(W_i * x + W_hi * h + b_i)
    
    # candidate cell state
    C_tilde = tanh_activation(W_c * x + W_hc * h + b_c)
    
    # cell state update 
    C = (f * C) + (i * C_tilde)
    
    # output gate
    o = sigmoid(W_o * x + W_ho * h + b_o)
    
    # hidden state update
    h = o * tanh_activation(C)
    
    print(f"--- Time Step t={t}, Input x_{t}={x} ---")
    print(f"f_{t} = {f:.3f}")
    print(f"i_{t} = {i:.3f}")
    print(f"C~_{t} = {C_tilde:.3f}")
    print(f"C_{t} = {C:.3f}")
    print(f"o_{t} = {o:.3f}")
    print(f"h_{t} = {h:.3f}\n")
```

```python
y_hat = W_y * h + b_y 

# final prediction 
print(f"Final Prediction y_hat = {y_hat:.3f}")
```
